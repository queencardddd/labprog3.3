```python
# Задание task_03_03_01

# Выполнила: Александрова Екатерина.
# Группа: ЦИБ - 251

"""
Ошибки (номера строк через пробел, данная строка — №2): 6 10 11 12 15
"""

def сумма_цифр(n):
    """Вернуть сумму цифр, меньших 5, для положительного целого числа `n`.
    Если таких цифр нет, вернуть 0.
    """

    c = 0
    while n > 0:
        цифра = n % 10

        if цифра < 5:
            c = c + цифра
        n //= 10
    return c

print(сумма_цифр(4834))
```
